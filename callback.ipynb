```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

```

Data generation


```python
x = np.arange (-80,81,0.5)
y = x ** 2 + 5 *x + 3
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
x_train.shape , x_test.shape
```

```python
Model1 = keras.Sequential()
Model1.add(keras.layers.Dense(units = 1024 , activation = "relu" , input_shape = [1]))
Model1.add(keras.layers.Dense(units = 1 ))
```

```python
Model1.compile(optimizer= "adam" , loss = "mean_squared_error")
```

```python
Model1.summary()
```

```python
hist_1 = Model1.fit(x_train , y_train, batch_size= 64 , epochs=1000 , validation_data = (x_test,y_test ))
```

```python
Model2 = keras.Sequential()
Model2.add (keras.layers.Dense(units= 256 , activation= "relu" , input_shape = [1]))
Model2.add (keras.layers.Dense(units= 256 , activation= "relu"))
Model2.add (keras.layers.Dense(units= 256 , activation= "relu"))
Model2.add (keras.layers.Dense(units= 256 , activation= "relu"))
Model2.add (keras.layers.Dense(units= 1))

```

```python
Model2.compile(optimizer= "adam" , loss = "mean_squared_error")
```

```python
Model2.summary()
```

```python
hist_2 = Model2.fit(x_train , y_train, batch_size= 64 , epochs=1000 , validation_data = (x_test,y_test ))
```

```python
predict1 = Model1.predict(x_test)
predict2 = Model2.predict(x_test)
```

```python
plt.scatter(x_test, y_test , color= "red"   ,linewidths= 4) # real data
plt.scatter(x_test,predict1, color= "blue"  ,linewidths= 2) # Model 1 with 2 layers
plt.scatter(x_test,predict2, color= "green" ,linewidths= 2) # Model2 with 4 Layres
plt.show()
```

```python
plt.plot(hist_1.history["val_loss"] [800:], color = "red")
plt.plot(hist_2.history["val_loss"] [800:], color = "blue")
plt.show()

```

```python
early_stop = keras.callbacks.EarlyStopping(monitor="val_loss", patience=50 , restore_best_weights=True)
```

```python
hist_3 = Model2.fit(x_train , y_train, batch_size= 128 , epochs=1000 , validation_data = (x_test,y_test ), callbacks=[early_stop])
```

```python
plt.plot(hist_3.history["val_loss"], color = "yellow")
plt.plot(hist_3.history["loss"], color = "red")

plt.show()
```
